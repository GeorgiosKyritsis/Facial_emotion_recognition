```python
import numpy as np
from model import model_generate
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
gender_table = {0: 'man',
                1: 'woman'
               }
```

```python
# Read the dataset
X_test = np.load('../../data/Gender/Test/final_data.npy')
Y_test = np.load('../../data/Gender/Test/final_labels_data.npy')
```

```python
model = model_generate(2)
```

```python
# Load the weights into our model
model.load_weights('../logs/weights-gender.h5')
```

### Examples

```python
img = np.reshape(X_test[10], newshape=(1,64,64,1))
imshow(np.reshape(img, newshape=(64,64)), cmap='gray')
```

```python
predicted_gender = gender_table[np.argmax(model.predict(img))]
print('Predicted Gender:', predicted_gender)
```
